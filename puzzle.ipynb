```python
c=0
for i in range(10,99):
    used=[]
    if((len(set(str(i)))==2) and int(str(i)[0]) not in used and int(str(i)[1]) not in used):
        used.append(int(str(i)[0]))
        used.append(int(str(i)[1]))
        for j in range(1,10):
            used=used[:2]
            if(j not in used):
                used.append(int(str(j)))
                mul=i*j
                if(len(str(mul))==2 and len(set(str(mul)))==2 and int(str(mul)[0]) not in used and int(str(mul)[1]) not in used):
                    used.append(int(str(mul)[0]))
                    used.append(int(str(mul)[1]))
                    for k in range(10,99):
                        used=used[:5]
                        if((len(set(str(k)))==2) and int(str(k)[0]) not in used and int(str(k)[1]) not in used):
                            add=mul+k
                            used.append(int(str(k)[0]))
                            used.append(int(str(k)[1]))
                            if((len(str(add))==2 and len(set(str(add)))==2) and int(str(add)[0]) not in used and int(str(add)[1]) not in used):
                                used.append(int(str(add)[0]))
                                used.append(int(str(add)[1]))
                                c+=1
                                
                                print("\n\n========================")
                                print("\nSolution : ",c,"\n")
                                
                                print(" ",i)
                                print('*',j)
                                print('----')
                                print(' ',mul)
                                print('+',k)
                                print('----')
                                print('  ',add)
                                
                                print("\nUnique Digits are = ",*used)
                                
                                print()
                   
```
